```python
def selectionsort(var):                                 # inisiasi nama fungsi
    for i in range(len(var)):                           # untuk setiap elemen/indeks dalam list
        min_idx = i                                     # inisiasi min_idx = i, untuk menampung nilai minimum
        for j in range(i +1, len(var)):                 # looping untung elemen yang tersisa dalam list, dimulai dari indeks i+1
            if var[min_idx]>var[j]:                     # dalam looping kedua, dibandingkan, jika nilai elemen min_idx > nilai elemenj
                min_idx =j                              # maka nilai min_idx diganti dengan j (untuk nilai terkecil)
        var[i], var[min_idx]= var[min_idx], var[i]      # setelah looping selesai, nilai elemen ditukar, jadi elemen ke i akan menjadi elemen terkecil
    return var          # return var
selectionsort([2,4,6,8,1])
```

```python
def sortganjilgenap(list):
    ganjil = []
    genap= []
    for angka in list:
        if angka%2==0:
            genap.append(angka)
        else:
            ganjil.append(angka)
    
    def sortingnya(list):
        iterasi = 0                               
        while iterasi < (len(list)-1):              
            cursor = 0                                 
            while cursor < (len(list)-1):           
                if list[cursor]>list[cursor+1]:       
                    list[cursor], list[cursor+1]= list[cursor+1], list[cursor] 
                cursor+=1 
            iterasi +=1 
        return list
    a = sortingnya(ganjil)
    b = sortingnya(genap)
    return a+b
sortganjilgenap([1,12,4,3,6,7,9,22])
```

```python
graf = {
    'Gresik': {'Surabaya': 9, 'Mojokerto':11},
    'Surabaya': {'Gresik': 9, 'Sidoarjo': 3},
    'Mojokerto': {'Gresik': 11, 'Sidoarjo':13},
    'Sidoarjo':{'Surabaya': 3, 'Mojokerto': 13, 'Pasuruan': 4},
    'Pasuruan': {'Sidoarjo': 4, 'Jember': 6, 'Malang': 7},
    'Jember': {'Pasuruan': 6, 'Malang': 15},
    'Malang': {'Jember': 15, 'Pasuruan': 7}

}
```

```python
def find_shortest_path_dis(graph, node_awal, node_akhir, path=[], jarak=0):
    path = path + [node_awal]
    if node_awal == node_akhir:
        return path, jarak
    if node_awal not in graph:
        return None, None
    
    shortest_path = None
    shortest_distance = None

    for node, panjang in graph[node_awal].items():
        if node not in path:
            new_path, new_distance = find_shortest_path_dis(graph, node, node_akhir, path, jarak + panjang)

            if new_path and (new_distance is not None) and (shortest_distance is None or new_distance < shortest_distance):
                shortest_path = new_path
                shortest_distance = new_distance

    return shortest_path, shortest_distance

```

```python
rute, jarak = find_shortest_path_dis(graf, 'Gresik', 'Sidoarjo', path=[], jarak=0)
print('Rute Terpendek= ', rute)
print('Jaraak Terdekat= ', jarak)
```

```python
def find_all_paths(graph, node_awal, node_akhir, path=[]):
    path= path +[node_awal]
    if node_awal == node_akhir:
        print("===== Jalur yang ditemukan:", path)
        return [path]

    if node_awal not in graph:
        return []
    paths = []

    for node in graph[node_awal]: 
        print('Path saat ini', path) 
        print(f'Node tujuan berikutnya adalah {node} dari node sebelumnya yaitu {node_awal}')
        
        
        if node not in path:
            newpaths =find_all_paths(graph, node, node_akhir, path) #fungsi rekursi
            for newpath in newpaths:
                paths.append(newpath)
    return paths
```

```python
hasil = find_all_paths(graf, 'Gresik', 'Malang', path=[])
```
