diekstrak dan dimasukkan ke dalam stack. Kemudian tiap karakter tersebut dikeluarkan dan

#### __penjelasan__
1. buat fungsi untuk membalik kata
    - inisiasi array kosong untuk stack dan kata yang telak dibalik
    - memasukkan setiap karakter dari kata ke stack
    - mengeluarkan isi stack menggunakan fungsi pop, lalu dimasukkan ke array balik
    - terus diulang sampai isi stack nya habis
    - mengembalikan balik -> kata yang sudah dibalik dengan menggunakan join supaya dari array ke kalimat biasa
2. buat fungsi untuk menggunakan
    - inisiasi input untuk kata yang ingin dibalik
    - print hasil yang merupakan hasil dari penggunaan fungsi membalik kata tadi

```python
def Pembalik(kata):                                 
    stack = []
    balik = []
    for i in kata:
        stack.append(i)
    for i in range(len(stack)):
        balik.append(stack.pop())
    return (''.join(balik))

def PembalikKata():
    kata = input('masukkan kata yang ingin dibalik')
    print('Kata =', kata)
    hasil = Pembalik(kata)
    print('Kata setelah dibalik =', hasil)

PembalikKata()
```

```python
antriMax= 5
antri =[]

def antrian():
    a = input('Masukkan daftar antrian= ').split(',')
    print('Jumlah antrian: ', len(a[:5]))
    for i in range(len(a)):
        if len(antri) < antriMax:
            antri.insert(i,a[i])
            print(a[i], 'masuk antrian')
        else:
            print(f'maaf {a[i]} antrian sudah penuh')

def keluarAntri():
    x = len(antri)
    while x > 0:
        out= antri.pop(0)
        print(f'{out} keluar antrian')
        print(', '.join(antri + ['Kosong'] * (5 - len(antri))))
        x -= 1
    print('Antrian kosong')
    print('Jumlah antrian: 0 orang')
    
print('>> Inputan data antrian')
antrian()
print('''
      ''')
print('>> Satu per satu keluar dari antrian')
keluarAntri()
```

Preorder traversal: ['Z', 'A', 'X', 'B', 'N', 'W', 'Q', 'G', 'R', 'F', 'S', 'D', 'K', 'J', 'P']

Inorder traversal: ['B', 'X', 'N', 'Q', 'W', 'A', 'Z', 'G', 'D', 'S', 'K', 'F', 'P', 'J', 'R']

Postorder traversal: ['B', 'N', 'Q', 'W', 'X', 'A', 'G', 'D', 'K', 'S', 'P', 'J', 'F', 'R', 'Z']

### __Penjelasan__
__Membuat 3 Fungsi: Postorder, Inorder, Preorder__
- membuat class baru yang bernama Node() dan memodifikasi variabel yang dimasukkan dengan "def init()"
- variabel memiliki 4 value

#### pre order
- membuat fungsi untuk melakukan penelusuran tree dengan pre-order traversal dengan input tree yang ingin dibaca dan traversal
- karena ini pre-order, maka semua value dari tree di tambahkan dahulu kedalam 1 string sehingga value pertamanya adalah yang paling atas
- setelah itu baru diurutkan valuenya secara preorder dimulai kiri kemudian tengah lalu kanan
- setelah selesai mengurutkan, maka fungsi akan mengeluarkan string hasil urutannya
#### inorder
- membuat fungsi untuk melakukan penelusuran tree dengan in-order traversal dengan input tree yang ingin dibaca dan traversal
- karena ini in-order maka kita harus mengurutkan valuenya dari kiridahulu sebelum ditambahkan kedalam 1 string karena kita akan mulai dari yang paling kiri
- setelah itu baru mengurutkan dari tengah kemudian kanan
- setelah selesai mengurutkan, maka fungsi akan mengeluarkan string hasil urutannya
#### postorder
- membuat fungsi untuk melakukan penelusuran tree dengan post-order traversal dengan input tree yang ingin dibaca dan traversal
- karena ini post-order dan kite ingin urutan dimulai dari yang paling bawah, maka langkah yang kita lakukan adalah mengurutkan valuenya semua
- semua value diurutkan terlebih dahulu dari kiri kemudian tengah lalu kanan sehingga hasil setelah pengurutan akan dimulai dari bawah
- setelah semua terurutkan baru kita menggabungkannya kedalam 1 string dan kemudian program akan mengeluarkan string hasil urutannya

__Memasukkan bentuk tree__
- membuat bentuk tree menggunakan class yang terlah ada sebelumnya
- untuk membuatnya kita harus menentukan arah root/akar agar bisa dimasukkan kedalam 'Node'-nya
- dalam membuatnya kita menyesuaikan variabel-variabel dengan apa yang ada di gambar/soal praktikum

__Print hasil dengan memanggil fungsi__
- Menggunakan output hasil berupa print untuk setiap metode



```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.mid = None
        self.right = None

class Tree:
    def __init__(self, root):
        self.root = Node(root)

    # Metode Prerder
    def preorder(self, start, result):
        if start:
            if start == self.root:
                print(f"{start.value} adalah root")
            elif not start.left and not start.mid and not start.right:
                print(f"{start.value} adalah leaf")
            result += (str(start.value))
            result = self.preorder(start.left, result)
            result = self.preorder(start.mid, result)
            result = self.preorder(start.right, result)
        return list(result)

    # Metode In-Order
    def inorder(self, start, result):
        if start:
            result = self.inorder(start.left, result)
            if start == self.root:
                print(f"{start.value} adalah root")
            elif not start.left and not start.mid and not start.right:
                print(f"{start.value} adalah leaf")
            result += str(start.value)
            result = self.inorder(start.mid, result)
            result = self.inorder(start.right, result)
        return list(result)


    def postorder(self, start, result):
        if start:
            result = self.postorder(start.left, result)
            result = self.postorder(start.mid, result)
            result = self.postorder(start.right, result)
            if start == self.root:
                print(f"{start.value} adalah root")
            elif not start.left and not start.mid and not start.right:
                print(f"{start.value} adalah leaf")
            result += str(start.value)
        return list(result)

```

```python
tree = Tree("Z")
tree.root.left = Node("A")
tree.root.mid = Node("G")
tree.root.right = Node("R")
tree.root.right.left = Node("F")                        #kan F cuma 1 cangan jdi pake left, bukan right, soalnya diitung left dulu
tree.root.right.left.left = Node("S")
tree.root.right.left.left.left = Node("D")
tree.root.right.left.left.mid = Node("K")
tree.root.right.left.mid = Node("J")                    # ini pake mid, soalnya di cbang utama kan ada 3, left mid sama right. J cabang kedua jadi pake mid
tree.root.right.left.mid.left = Node("P")
tree.root.left.left = Node("X")
tree.root.left.left.left = Node("B")
tree.root.left.left.mid = Node("N")
tree.root.left.left.right = Node("W")               # ini pake right soalnya dia cabang ke 3 jdi pas urutannya left, mid, right
tree.root.left.left.right.left = Node("Q")
```

```python
# Memanggil metode preorder
print("Preorder traversal:", tree.preorder(tree.root, ""))
```

```python
# Memanggil metode inorder
print("Inorder traversal:", tree.inorder(tree.root, ""))
```

```python
# Memanggil metode postorder
print("Postorder traversal:", tree.postorder(tree.root, ""))
```
