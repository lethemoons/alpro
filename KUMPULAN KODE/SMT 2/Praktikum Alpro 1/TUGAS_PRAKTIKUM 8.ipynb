a. fungsi untuk mengambil elemen kedua dari terakhir pada list

```python
def terakhir(list):      # inisiasi fungsi
    return list[-2]      # mengambil element kedua dari belakang dengan menggunakan list[-2] karena yang diambil element urutan 2 dari belakang

terakhir([1,2,34,5,6])   # panggil fungsi
```

b. Fungsi untuk mengkuadratkan setiap elemen pada list yang berisi angka.

```python
def kuadrat(list):          # inisiasi fungsi
    baru =[]                # membuat list kosong
    for i in list:          # untuk setiap element dari list 
        baru.append(i*i)    # setiap element dikali element itu sendiri dan dimasukkan ke list kosong
    return baru             # return list kosong tadi yang sudah terisi bilangan kuadrat dari masing-masing element

kuadrat([1,2,3,4,5])        # panggil fungsi
```

```python
def yesno(file):                                        # inisiasi fungsi
    #asli = file                                         # 
    ganti= {'No': 'yes', 'no':'yes', 'NO': 'yes'}        # membuat dictionary untuk berbagai macam jenis no, pakai dict supaya nnti selain kata no format filenya tidak berubah. kalau pakai lower nnti isi filenya jdi huruf kecil semua
    for lama, baru in ganti:                             # untuk key, value di dict ganti
        diganti = file.replace(lama,baru)                # replace file, lama dengan baru
    return diganti                                       # return diganti

yesno(moby())               # panggil fungsi dengan menggunakan parameter moby()
```

```python
def selection_sort_asc(arr):
    n = len(arr)
    for i in range(n):
        max_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[max_idx]:
                max_idx = j
        arr[i], arr[max_idx] = arr[max_idx], arr[i]
    return arr

def sortmoby(file):
   a= file.split()
   return selection_sort_asc(a)

mob = """ The pale Usher--threadbare in coat, heart, body, and brain; I see him
now. He was ever dusting his old lexicons and grammars, with a queer
handkerchief, mockingly embellished with all the gay flags of all
the known nations of the world. He loved to dust his old grammars; it
somehow mildly reminded him of his mortality."""
sortmoby(mob)
```

d. Fungsi untuk mengecek isi dari sebuah list (ada atau tidak)

```python
def cekisi(list):               # inisiasi fungsi
    if len(list) !=0:           # menggunakan len(list), jika != 0 -> berarti list ada isinya
        return 'ADA'            # return ada
    else:                       # else berarti len(list) == 0 -> list kosong
        return 'TIDAK ADA'      # return tidak ada
    
print(cekisi([-9,0,8]))
print(cekisi([]))
```

a. Fungsi kalkulator() berisi program dengan meminta pengguna untuk memilih
operasi matematika yang diinginkan (+, -, /, *)

```python
def kalkulator():                                                 # inisiasi fungsi
    operator = input('pilih operasi +,-,/,*')                     # membuat input untuk pemilihan operator
    while operator not in (['+', '-', '*', '/']):                 # jika operator yang diinputkan tidak termasuk di ['+', '-', '*', '/'] maka akan print error
        print('error, silakan masukkan operator yang sesuai')       # jika operator tidak sesuai maka print error dan user disuruh menginputkan operator yang benar
        operator = input('pilih operasi +,-,/,*')                   # input ulang operator, loop akan terus berjalan sampai user memasukkan operator yang sesuai
        
    num1= int(input('masukkan angka'))                              # jika operator sudah sesuai maka lanjut untuk input angka pertama 
    num2= int(input('masukkan angka'))                              # input angka kedua
    
    if operator =='+':                                          # jika operator sama dengan +
        return num1 + num2                                      # maka return num1+num2
    elif operator =='*':                                        # jika operator sama dengan *
        return num1 * num2                                      # maka return num1*num2
    elif operator == '/' and num2 == 0:                         # jika operator sama dengan / dan num 2 sama dengan 0
        return 'tidak bisa dibagi dengan 0!'                    # maka return tidak bisa dibagi dengan 0
    elif operator == '/':                                       # jika operator sama dengan /
        return num1 / num2                                      # maka return num1/num2
    elif operator == '-':                                       # jika operator sama dengan -
        return num1-num2                                        # maka return num1-num2
    else:                                                       # else, return error
        return 'error'                                          #
    
kalkulator()                        # panggil fungsi
     
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_moby_dick():
  """Scrapes the Moby Dick text from the Gutenberg repository.

  Returns:
      str: The scraped Moby Dick text.
  """

  # Target URL for Moby Dick
  url = "https://www.gutenberg.org/files/5228/5228-h/5228-h.htm"

  # Fetch the web page content
  response = requests.get(url)

  # Check for successful response
  if response.status_code == 200:
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract text from paragraph elements
    moby_dick_text = ""
    for paragraph in soup.find_all('p'):
      # Clean text and remove extra spaces
      text = paragraph.text.strip()
      moby_dick_text += text + "\n\n"  # Add double newline for better readability

    return moby_dick_text
  else:
    print(f"Error: Failed to download Moby Dick text. Status code: {response.status_code}")
    return None

# Scrape and print the Moby Dick text
moby_dick_text = scrape_moby_dick()

if moby_dick_text:
  print(moby_dick_text)
else:
  print("Moby Dick text scraping failed.")
```
